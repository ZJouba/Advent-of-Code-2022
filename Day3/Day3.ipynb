```python
import requests

file = requests.get('https://raw.githubusercontent.com/ZJouba/Advent-of-Code-2022/main/Day3/Day3.txt')
input = list(filter(None, file.text.split('\n')))
input2 = input
input = list(map(lambda x: [x[0:round(len(x)/2)], x[round(len(x)/2):len(x)]], input))
print(input)
```

```python
def check(pack):
  packSet = set(pack[0])
  letter = [x for x in pack[1] if x in packSet][0]
  value = (ord(letter) & 31)
  return int(value + 26 if letter.isupper() else value)


values = list(map(lambda pack: check(pack), input))
print(sum(values))
```

```python
from itertools import zip_longest

def grouper(n, iterable, fillvalue=None):
    args = [iter(iterable)] * n
    return zip_longest(fillvalue=fillvalue, *args)

def checker(pack):
  packSet = set(pack[0])
  packSet2 = set(pack[1])
  letter = [x for x in pack[2] if x in packSet and x in packSet2][0]
  value = (ord(letter) & 31)
  return int(value + 26 if letter.isupper() else value)

groups = list(grouper(3, input2))
print(groups)
print(sum(list(map(lambda pack: checker(pack), groups))))
```
