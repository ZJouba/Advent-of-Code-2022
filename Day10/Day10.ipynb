```python
def nthStep(step):
  pos = len([step - item[2] for item in input if step - item[2] >= 0])
  
  if input[pos-1][2] == step and input[pos-1][0] == 'addx':
    pos -= 1
  elif input[pos][0] != 'addx':
    pos += 1

  # print(input[:pos])

  ignore = list(filter(lambda i: i[0] != 'noop', input[:pos]))
  ignoreTotal = sum([i[1] for i in ignore])

  print(ignoreTotal + 1)
  return (ignoreTotal + 1)

# nthStep(220)
print(sum([nthStep(i) * i for i in [20, 60, 100, 140, 180, 220]]))
```

```python
from copy import deepcopy
import numpy as np
np.set_printoptions(threshold=np.inf, linewidth=500)

x = 1
program = []
c_input = deepcopy(input)
i = 1
row = 0

while len(c_input) > 0:
  cmd = c_input.pop(0)
  if cmd[0] == 'addx':
    for w in range(2):
      i += w
      if i > 39:
        i = 0
        row += 1
      program.append([x, i, row])
    x += cmd[1]
  else:
    program.append([x, i, row])
  
  if i > 39:
    i = 0
    row += 1
  i += 1

print(program)
overlap = [c for c in program if c[0] in range(c[1] - 2, c[1] + 1)]
print(overlap)

screen = np.tile('.', [6, 40])
for p in overlap:
  screen[p[2], p[1]-1] = '#'

# print(screen)
print('\n'.join(''.join(str(cell) for cell in row) for row in screen))
```
